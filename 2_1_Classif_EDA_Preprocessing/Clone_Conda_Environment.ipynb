Before any major or uncertain update (like installing a new library that you are not familiar with), it is a good practice to backup your conda environment. This way, you can always go back to a previous state if something goes wrong. The easiest way to do that is to clone your environment. That way, if anything goes wrong, you can always go back to the previous state. To clone the environment, you can open a terminal and use the following command:


```python
conda create --name MLMIC25_bkup_20250116 --clone MLMIC25
```


The name we have chosen for the clone in this example is `MLMIC25_bkup_20250116`. The date is included in the name to make it easier to identify the backup, but you could also include something that reminds you of the reason that triggered the backup. 

Then, if anyhting goes wrong, you can always go back to the previous state by:

+ checking that the backup environment is working by activating it with the following command:


```python
conda activate MLMIC25_bkup_20250116
```

+ deleting the current environment with the following command:

```python
conda remove --name MLMIC25 --all
```

+ Reverse the clone process by creating a new environment with the same name as the original one. This should bring you back to the previous state:

```python
conda create --name MLMIC25 --clone MLMIC25_bkup_20250116
```

Remember that you can always check the list of environments with the following command:

```python
conda env list
```


When I am working on a project, I usually keep a sort of minimal backup environment for the initial setup once I get it to work. And then I create backups every time I am going to do something that could potentially break the environment. If the change is successful, I delete the backup after a few sessions of work. By keeping some notes about the backups schedule and the changes involved, I can always go back to a previous safe and stable state.

For this and other conda environment management operations we recommend that you visit conda's website and check the documentation at this url:

[Managing environments](https://docs.conda.io/projects/conda/en/latest/user-guide/tasks/manage-environments.html#)
